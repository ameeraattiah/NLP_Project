```python
import PyPDF2

def read_pdf(file_path):
    try:
        # Open the PDF file
        with open(file_path, 'rb') as pdf_file:
            # Create a PDF Reader object
            reader = PyPDF2.PdfReader(pdf_file)
            
            # Read each page and print its content
            for page_number, page in enumerate(reader.pages):
                print(f"--- Page {page_number + 1} ---")
                print(page.extract_text())
                print("\n")
    except Exception as e:
        print(f"Error reading the PDF file: {e}")

# Example usage
file_path = "./Journals/القرينة النحوية وأمن اللبس.pdf"  # Replace with the path to your PDF
read_pdf(file_path)
```

```python
from pdf2image import convert_from_path
import pytesseract
from PIL import Image

# Path to Tesseract OCR executable (Update path for your system)
pytesseract.pytesseract.tesseract_cmd = r'/opt/homebrew/bin/tesseract'

# Function to perform OCR on an Arabic PDF
def read_arabic_pdf(pdf_path, language="ara", poppler_path="/opt/homebrew/bin/"):
    try:
        # Convert PDF pages to images
        images = convert_from_path(pdf_path, poppler_path=poppler_path)
        
        # Extract text from each image
        full_text = ""
        for page_number, image in enumerate(images, start=1):
            # print(f"Processing page {page_number}...")
            text = pytesseract.image_to_string(image, lang=language)
            full_text += f"\n--- Page {page_number} ---\n{text}"
        
        return full_text
    except Exception as e:
        print(f"Error reading PDF: {e}")
        return None

# Path to your Arabic PDF file
pdf_path = "./Journals/القرينة النحوية وأمن اللبس.pdf"

# Read the Arabic text
arabic_text = read_arabic_pdf(pdf_path)

# Print the extracted text
if arabic_text:
    print(arabic_text)
```

```python
import os
import re
import csv
import json
from pdf2image import convert_from_path
import pytesseract
from PIL import Image
from datasketch import MinHash, MinHashLSH
from nltk.tokenize import sent_tokenize
import fasttext

# Load FastText model
FASTTEXT_MODEL_PATH = "lid.176.bin"  # Update with the path to your FastText model
language_model = fasttext.load_model(FASTTEXT_MODEL_PATH)

# Path to Tesseract OCR executable
pytesseract.pytesseract.tesseract_cmd = r'/opt/homebrew/bin/tesseract'

# OCR Function
def read_arabic_pdf(pdf_path, language="ara", poppler_path="/opt/homebrew/bin/"):
    """Read Arabic text from a PDF using Tesseract OCR."""
    try:
        images = convert_from_path(pdf_path, poppler_path=poppler_path)
        full_text = ""
        for page_number, image in enumerate(images, start=1):
            text = pytesseract.image_to_string(image, lang=language)
            full_text += f" {text}"  # Removed `\n` between pages
        return full_text
    except Exception as e:
        print(f"Error reading PDF {pdf_path}: {e}")
        return None

# Utility Functions
def detect_language(text):
    """Detect the primary language of the text."""
    cleaned_text = text.replace("\n", " ").strip()  # Remove newlines
    prediction = language_model.predict(cleaned_text[:1000])  # Process only the first 1000 characters
    language = prediction[0][0].replace("__label__", "")
    confidence = prediction[1][0]
    return language, confidence

def normalize_and_validate_text(text):
    """Normalize Arabic text, remove unbalanced brackets, and clean up."""
    # Normalize Arabic text
    text = re.sub(r"[ًٌٍَُِّْ]", "", text)  # Remove Arabic diacritics
    text = text.replace("أ", "ا").replace("إ", "ا").replace("آ", "ا")
    text = text.replace("\u200f", "")  # Remove U+200F
    text = text.replace("\n", " ")  # Remove newlines for better readability

    # Remove unbalanced brackets
    bracket_pairs = {"(": ")", "[": "]", "{": "}"}
    for open_bracket, close_bracket in bracket_pairs.items():
        while text.count(open_bracket) > text.count(close_bracket):
            text = text[:text.rfind(open_bracket)] + text[text.rfind(open_bracket) + 1:]
        while text.count(close_bracket) > text.count(open_bracket):
            text = text[:text.find(close_bracket)] + text[text.find(close_bracket) + 1:]

    # Remove reference-like patterns
    text = re.sub(r"\[\d+\]", "", text)  # Remove patterns like [1], [23]
    text = re.sub(r"^\d+\.", "", text, flags=re.MULTILINE)  # Remove lines starting with 1., 2., etc.

    return text.strip()

def deduplicate_sentences(text):
    """Remove duplicate sentences within a single document."""
    sentences = text.split(". ")
    unique_sentences = list(dict.fromkeys(sentences))  # Preserve order
    return ". ".join(unique_sentences)

def deduplicate_documents(data, threshold=0.8):
    """Remove duplicate documents using MinHash."""
    lsh = MinHashLSH(threshold=threshold)
    unique_data = []
    for idx, record in enumerate(data):
        # Concatenate tokenized sentences into a single string for deduplication purposes
        text = " ".join(record['text'])  # Join list of sentences into a single string
        tokens = text.split()  # Split into tokens
        m = MinHash()
        for token in tokens:
            m.update(token.encode("utf-8"))
        if not any(lsh.query(m)):
            lsh.insert(str(idx), m)
            unique_data.append(record)
    return unique_data

def process_pdfs(folder_path, output_file="output.csv"):
    """Process all PDF files in a folder and save results as a CSV."""
    processed_data = []

    for file_name in os.listdir(folder_path):
        if file_name.endswith(".pdf"):
            file_path = os.path.join(folder_path, file_name)
            print(f"Processing file: {file_name}")

            try:
                # Perform OCR on the PDF
                extracted_text = read_arabic_pdf(file_path)

                # Remove non-Arabic text
                sentences = extracted_text.split(". ")
                arabic_sentences = [
                    sentence for sentence in sentences if detect_language(sentence)[0] == "ar"
                ]
                arabic_text = ". ".join(arabic_sentences)

                # Normalize and validate text
                normalized_text = normalize_and_validate_text(arabic_text)

                # Deduplicate sentences
                deduplicated_text = deduplicate_sentences(normalized_text)

                # Skip if no valid Arabic text remains
                if not deduplicated_text.strip():
                    continue

                # Tokenize into sentences
                tokenized_sentences = sent_tokenize(deduplicated_text)

                # Detect language for metadata
                language, confidence = detect_language(deduplicated_text)

                # Metadata for the document
                metadata = {
                    "Document Name": file_name,
                    "Language": language,
                    "Language Confidence": round(confidence, 2),
                    "Token Count": len(deduplicated_text.split())
                }

                # Save processed document with metadata
                processed_data.append({
                    "text": tokenized_sentences,
                    "metadata": metadata
                })

            except Exception as e:
                print(f"Error processing file {file_name}: {e}")
                continue

    # Deduplicate documents
    unique_data = deduplicate_documents(processed_data)

    # Save all processed data to a CSV file
    with open(output_file, "w", encoding="utf-8", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Text (Tokenized Sentences)", "Metadata"])
        for record in unique_data:
            writer.writerow([
                "\n".join(record["text"]),  # Join sentences for readability
                json.dumps(record["metadata"], ensure_ascii=False, indent=4)  # Pretty JSON format
            ])

    print(f"Processed {len(unique_data)} unique documents successfully.")
    return output_file

# Example usage
folder_path = "./Journals"  # Folder containing PDF files
output_file = "processed_output2.csv"  # Output CSV file
process_pdfs(folder_path, output_file)
print(f"Output saved to {output_file}")
```
